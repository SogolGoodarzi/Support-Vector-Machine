# **Support Vector Machine**

```python
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
from mlxtend.plotting import plot_decision_regions

iris = datasets.load_iris()
data = iris.data[:, :2] #data
label = iris.target #label

# Create the SVM
svm = LinearSVC()
OVRclassifier = OneVsRestClassifier(svm)
OVRclassifier = OVRclassifier.fit(data, label)

# Plotting decision boundary
plot_decision_regions(data, label, clf = OVRclassifier, legend=2)
plt.show()
```

######**Accuracy of train data, Confusion Matrix, Confidence Matrix**

```python
from sklearn import metrics
## Accuracy of train data:
label_predicted = OVRclassifier.predict(data) 
Accuracy = metrics.accuracy_score(label, label_predicted) 
print('Accuracy of train data is: {}'.format(Accuracy))

## Confusion Matrix:
conf_mat = metrics.confusion_matrix(label, label_predicted)
conf_mat = conf_mat/sum(conf_mat[0,:])   ## changing the scale of matrix
display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_mat, display_labels = [0,1,2])
display.plot()
plt.title('Confusion Matrix')
plt.show()

## Confidence Matrix:
for i in range(len(conf_mat)):
  conf_mat[:,i] = conf_mat[:,i]/sum(conf_mat[:,i])
display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_mat, display_labels = [0,1,2])
display.plot()
plt.title('Confidence Matrix')
plt.show()
```
